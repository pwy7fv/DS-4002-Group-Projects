```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the time periods for comparison
time_periods = {
    "Fast and Furious 1": [(2001, 2001), (2011, 2025)],
    "Fast and Furious 2": [(2003, 2003), (2013, 2025)],
    "Fast and Furious 3": [(2006, 2006), (2016, 2025)],
    "Fast and Furious 4": [(2009, 2009), (2019, 2025)],
    "Fast and Furious 5": [(2011, 2011), (2021, 2025)],
    "Fast and Furious 6": [(2013, 2013), (2023, 2025)],
}

# List to store p-values from the t-tests
p_values = []

# Loop through each movie and calculate the t-test for the two periods
for movie, periods in time_periods.items():
    period_1, period_2 = periods

    # Filter the dataset for the current movie
    filtered_df = combined_FF[combined_FF["Movie"] == movie]

    # Separate the data into two periods based on the time ranges
    period_1_df = filtered_df[filtered_df["Year"] >= period_1[0]]
    period_1_df = period_1_df[period_1_df["Year"] <= period_1[1]]

    period_2_df = filtered_df[filtered_df["Year"] >= period_2[0]]
    period_2_df = period_2_df[period_2_df["Year"] <= period_2[1]]

    # Perform the t-test (assuming normality of sentiment scores in each period)
    t_stat, p_value = stats.ttest_ind(period_1_df['Overall Percentage Score'], period_2_df['Overall Percentage Score'])

    # Append the p-value for this movie to the list
    p_values.append(p_value)

# Create a bar plot showing the p-values from the t-tests for each movie
plt.figure(figsize=(10, 6))
plt.bar(time_periods.keys(), p_values, color='skyblue')

# Add labels and title
plt.xlabel('Movies')
plt.ylabel('P-value')
plt.title('P-values from T-tests Comparing Sentiment Scores for Two Time Periods')

# Display the p-values on top of each bar for better readability
for i, p_value in enumerate(p_values):
    plt.text(i, p_value + 0.01, f'{p_value:.4f}', ha='center', va='bottom', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(time_periods.keys(), p_values, color='skyblue', s=100)  # 's' controls size of dots
plt.axhline(y=0.05, color='red', linestyle='--', label='Significance Threshold (p = 0.05)')

# Add labels and title
plt.xlabel('Movies')
plt.ylabel('P-value')
plt.title('P-values from T-tests Comparing Sentiment Scores for Two Time Periods')

# Annotate p-values above the dots
for i, p_value in enumerate(p_values):
    plt.text(i, p_value + 0.01, f'{p_value:.4f}', ha='center', va='bottom', fontsize=10)

# Display the threshold line and legend
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the time periods for each movie, used CHAT GPT here
time_periods = {
    "Fast and Furious 1": [(2001, 2001), (2011, 2025)],
    "Fast and Furious 2": [(2003, 2003), (2013, 2025)],
    "Fast and Furious 3": [(2006, 2006), (2016, 2025)],
    "Fast and Furious 4": [(2009, 2009), (2019, 2025)],
    "Fast and Furious 5": [(2011, 2011), (2021, 2025)],
    "Fast and Furious 6": [(2013, 2013), (2023, 2025)],
}

# List to store the results for plotting
data_to_plot = []

# Loop through each movie and calculate the average rating for the two periods, Used CHAT GPT here
for movie, periods in time_periods.items():
    period_1, period_2 = periods

    # Filter the dataset for the current movie
    filtered_df = combined_FF[combined_FF["Movie"] == movie]

    # Separate the data into two periods based on the time ranges
    period_1_df = filtered_df[filtered_df["Year"] >= period_1[0]]
    period_1_df = period_1_df[period_1_df["Year"] <= period_1[1]]

    period_2_df = filtered_df[filtered_df["Year"] >= period_2[0]]
    period_2_df = period_2_df[period_2_df["Year"] <= period_2[1]]

    # Calculate the average rating for each period
    avg_rating_1 = period_1_df["Rating"].mean()
    avg_rating_2 = period_2_df["Rating"].mean()

    # Add the data to the list for plotting, used CHAT GPT to figure out how to seperate by time period
    data_to_plot.append({"Movie": movie, "Time Period": f"{period_1[0]}-{period_1[1]}", "Average Rating": avg_rating_1})
    data_to_plot.append({"Movie": movie, "Time Period": f"{period_2[0]}-{period_2[1]}", "Average Rating": avg_rating_2})

# Convert the list to a DataFrame
plot_df = pd.DataFrame(data_to_plot)

# Create a grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x="Movie", y="Average Rating", hue="Time Period", data=plot_df, palette="Set2")

# Add labels and title
plt.xlabel('Movies')
plt.ylabel('Average Rating')
plt.title('Average Rating for Each Fast and Furious Movie by Time Period')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
