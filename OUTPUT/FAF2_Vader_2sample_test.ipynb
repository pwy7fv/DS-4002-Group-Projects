Vader analysis performed using code from  cited reference [13] in listed references in README


```python
#Got this code from cited reference [13] in listed references in README
def sentiment_scores(index, row):
    # Extract Content, Author Name, Date, Rating
    sentence = row["Content"]
    name = row["Author Name"]
    title = row['Title']
    review_date = row["Date"]
    review_year = row["Year"]
    rating = row["Rating"]

    # Create a SentimentIntensityAnalyzer object
    sid_obj = SentimentIntensityAnalyzer()

    # Get sentiment scores
    sentiment_dict = sid_obj.polarity_scores(sentence)

    # Determine overall sentiment
    if sentiment_dict['compound'] >= 0.05:
        sentiment_label = "Positive"
    elif sentiment_dict['compound'] <= -0.05:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"

    # Return sentiment results as a Series (so it's added as columns in the original DataFrame)
    return pd.Series({
        "Sentiment Overall": sentiment_label,
        "Negative Score": sentiment_dict['neg'],
        "Neutral Score": sentiment_dict['neu'],
        "Positive Score": sentiment_dict['pos'],
        "Overall Percentage Score": sentiment_dict['compound']
    })
```

```python
df3_03[["Sentiment Overall", "Negative Score", "Neutral Score", "Positive Score", "Overall Percentage Score"]] = df3_03.apply(lambda row: sentiment_scores(row.name, row), axis=1)
```

```python
import numpy as np
import scipy.stats as stats

# Extract the overall sentiment scores from both datasets
sentiment_03 = np.array(df3_03["Overall Percentage Score"])
sentiment_13 = np.array(df2_13["Overall Percentage Score"])

var_03 = np.var(sentiment_03, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)
var_13 = np.var(sentiment_13, ddof=1)

print(f"Variance of 2009 Sentiment Scores: {var_03:.4f}")
print(f"Variance of 2019-2025 Sentiment Scores: {var_13:.4f}")

t_stat, p_value = stats.ttest_ind(sentiment_03, sentiment_13, equal_var=False)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in sentiment between 2001 and 2011-2025.")
else:
    print("Fail to reject the null hypothesis: No significant difference in sentiment.")

```
