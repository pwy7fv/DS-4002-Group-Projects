```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#creating dataframe of FF3_vader_2006 data to analyze
df3_vader06 = pd.read_csv('/content/DS-4002-Group-Projects/DATA/Cleaned Data/FF3_Vader2006.csv', low_memory=False)
df3_vader06.head()
```

```python
#creating dataframe of FF3_vader_dec data to analyze(2016 - 2025)
df3_vader_dec = pd.read_csv('/content/DS-4002-Group-Projects/DATA/Cleaned Data/FF3_Vader_dec.csv', low_memory=False)
df3_vader_dec.head()
```

```python
# Group by the 'Year' column and calculate the mean of 'Overall Percentage Score'
avg_sentiment_06 = df3_vader06['Overall Percentage Score'].mean()
avg_sentiment_dec = df3_vader_dec['Overall Percentage Score'].mean()

print(f"Average sentiment score for 2006: {avg_sentiment_06}")
print(f"Average sentiment score for 2016 - 2025: {avg_sentiment_dec}")
```

**Vizulations of Vader Analysis for FF3 2006 and FF3 2016 - 2025**

```python
import matplotlib.pyplot as plt

# Plotting histograms for both periods
plt.figure(figsize=(12, 6))

# Histogram for 2006
plt.subplot(1, 2, 1)
plt.hist(df3_vader06['Overall Percentage Score'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Sentiment Distribution (2006)')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')

# Histogram for 2016-2025
plt.subplot(1, 2, 2)
plt.hist(df3_vader_dec['Overall Percentage Score'], bins=20, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Sentiment Distribution (2016-2025)')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Plotting box plots for both periods
plt.figure(figsize=(8, 6))

# Box plot for 2006
plt.boxplot(df3_vader06['Overall Percentage Score'], positions=[1], widths=0.5, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='black'), medianprops=dict(color='black'))

# Box plot for 2016-2025
plt.boxplot(df3_vader_dec['Overall Percentage Score'], positions=[2], widths=0.5, patch_artist=True,
            boxprops=dict(facecolor='salmon', color='black'), medianprops=dict(color='black'))

plt.xticks([1, 2], ['2006', '2016-2025'])
plt.title('Sentiment Distribution by Year Range')
plt.ylabel('Sentiment Score')

plt.show()
```

```python
# Count the number of positive, negative, and neutral reviews for both periods
sentiment_counts_06 = df3_vader06['Overall Sentiment'].value_counts()
sentiment_counts_dec = df3_vader_dec['Overall Sentiment'].value_counts()

# Convert the counts to dataframes for easy plotting
sentiment_counts_06_df = sentiment_counts_06.reset_index()
sentiment_counts_06_df.columns = ['Sentiment', 'Count']
sentiment_counts_06_df['Period'] = '2006'

sentiment_counts_dec_df = sentiment_counts_dec.reset_index()
sentiment_counts_dec_df.columns = ['Sentiment', 'Count']
sentiment_counts_dec_df['Period'] = '2016-2025'

# Plot for 2006 reviews
plt.figure(figsize=(8, 5))
sns.barplot(x='Sentiment', y='Count', data=sentiment_counts_06_df, palette='Set2')
plt.title('Sentiment Distribution in 2006 Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Count of Reviews')
plt.show()

# Plot for 2016-2025 reviews
plt.figure(figsize=(8, 5))
sns.barplot(x='Sentiment', y='Count', data=sentiment_counts_dec_df, palette='Set2')
plt.title('Sentiment Distribution in 2016-2025 Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Count of Reviews')
plt.show()
```

```python
# Count sentiment occurrences and convert to percentages
sentiment_06 = df3_vader06["Overall Sentiment"].value_counts(normalize=True) * 100  # Convert to percentage
sentiment_dec = df3_vader_dec["Overall Sentiment"].value_counts(normalize=True) * 100  # Convert to percentage

# Pie chart labels and colors
labels = ["Positive", "Neutral", "Negative"]
colors = ["#2ecc71", "#e74c3c", "#f1c40f"]  # Green, Red, Yellow

# Create subplots for side-by-side pie charts
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for 2006
axes[0].pie(sentiment_06, labels=sentiment_06.index, autopct='%1.1f%%', colors=colors, startangle=140)
axes[0].set_title("Sentiment Distribution (2006)")

# Pie chart for 2016-2025
axes[1].pie(sentiment_dec, labels=sentiment_dec.index, autopct='%1.1f%%', colors=colors, startangle=140)
axes[1].set_title("Sentiment Distribution (2016-2025)")

# Show the pie charts
plt.tight_layout()
plt.show()
```

```python
# Concatenate both datasets into a single DataFrame
df3_combined = pd.concat([df3_vader06, df3_vader_dec], ignore_index=True)

# 2. Group by Year, Sentiment Overall, and Period
sentiment_counts_combined = df3_combined.groupby(["Year", "Overall Sentiment"]).size().reset_index(name="Count")

# 3. Pivot the data to get Sentiment as columns
sentiment_pivot_combined = sentiment_counts_combined.pivot_table(index=["Year"],
                                                                columns="Overall Sentiment",
                                                                values="Count",
                                                                aggfunc="sum",
                                                                fill_value=0).reset_index()

# 4. Plotting
plt.figure(figsize=(12, 6))

# Create a bar plot with 'Year' on the x-axis, and sentiment counts on the y-axis
sentiment_pivot_combined.plot(kind="bar", x="Year", stacked=True, colormap="viridis", figsize=(12, 6), ax=plt.gca())

# Labels and Title
plt.xlabel("Year")
plt.ylabel("Number of Reviews")
plt.title("Sentiment Distribution of Reviews by Year (2006 vs 2016-2025)")
plt.legend(title="Sentiment", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```
