```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
#creating dataframe of FF3_clean_2006 data to analyze
df3_06 = pd.read_csv('/content/DS-4002-Group-Projects/DATA/Cleaned Data/FF3_clean2006.csv', low_memory=False)
df3_06.head()
```

```python
#Got this code from cited reference [13] in listed references in README
def sentiment_scores(index, row):
    # Extract Content, Author Name, Date, Rating
    sentence = row["Content"]
    name = row["Author Name"]
    title = row['Title']
    review_date = row["Date"]
    review_year = row["Year"]
    rating = row["Rating"]

    # Create a SentimentIntensityAnalyzer object
    sid_obj = SentimentIntensityAnalyzer()

    # Get sentiment scores
    sentiment_dict = sid_obj.polarity_scores(sentence)

    # Determine overall sentiment
    if sentiment_dict['compound'] >= 0.05:
        sentiment_label = "Positive"
    elif sentiment_dict['compound'] <= -0.05:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"

    # Return sentiment results as a Series (so it's added as columns in the original DataFrame)
    return pd.Series({
        "Sentiment Overall": sentiment_label,
        "Negative Score": sentiment_dict['neg'],
        "Neutral Score": sentiment_dict['neu'],
        "Positive Score": sentiment_dict['pos'],
        "Overall Percentage Score": sentiment_dict['compound']
    })
```

```python
# Apply the function to the DataFrame and add new columns
df3_06[["Sentiment Overall", "Negative Score", "Neutral Score", "Positive Score", "Overall Percentage Score"]] = df3_06.apply(lambda row: sentiment_scores(row.name, row), axis=1)
```

**T-Test**

```python
import numpy as np
import scipy.stats as stats

# Extract the overall sentiment scores from both datasets
sentiment_06 = np.array(df3_06["Overall Percentage Score"])  # Ensure df3_2006 is loaded
sentiment_dec = np.array(df3_dec["Overall Percentage Score"])  # Ensure df3_2016 is loaded

# Print variance of both groups (checking assumption of equal variance)
var_06 = np.var(sentiment_06, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)
var_dec = np.var(sentiment_dec, ddof=1)

print(f"Variance of 2006 Sentiment Scores: {var_06:.4f}")
print(f"Variance of 2016-2025 Sentiment Scores: {var_dec:.4f}")

# Perform the two-sample t-test (Welch’s t-test since variances might be unequal)
t_stat, p_value = stats.ttest_ind(sentiment_06, sentiment_dec, equal_var=False)

# Display results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in sentiment between 2006 and 2016-2025.")
else:
    print("Fail to reject the null hypothesis: No significant difference in sentiment.")
```
